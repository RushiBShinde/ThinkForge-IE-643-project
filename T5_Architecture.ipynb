```python
import torch
from transformers import T5Tokenizer, T5ForConditionalGeneration

# 1. Load the tokenizer and the correct model for generation
# We'll use 't5-small' as it's quick to load.
model_name = 't5-small'
tokenizer = T5Tokenizer.from_pretrained(model_name)
model = T5ForConditionalGeneration.from_pretrained(model_name)

# 2. Define your task and input text
# The 'translate English to German: ' prefix tells T5 what task to perform.
task_prefix = ' English to Italian:'
input_text = 'That is a wonderful house.'

# 3. Tokenize the input
# The tokenizer converts your text into numerical IDs that the model can understand.
# return_tensors="pt" creates PyTorch tensors.
inputs = tokenizer(task_prefix + input_text, return_tensors="pt")

# 4. Generate the output from the model
# model.generate() is the core function for this.
output_sequences = model.generate(input_ids=inputs['input_ids'])

# 5. Decode the output
# The model outputs token IDs, so we use the tokenizer to convert them back to human-readable text.
# skip_special_tokens=True removes tokens like </s> (end of sentence).
decoded_output = tokenizer.decode(output_sequences[0], skip_special_tokens=True)

# 6. Print the results!
print("Input Text:", input_text)
print("Generated Output:", decoded_output)
```

# T5 for Ranking





```python
import torch

candidates = [
    "The Eiffel Tower is in Paris.",
    "The Eiffel Tower is not in India but in Paris",

    "It was built in 1889."

]
question = "Where is the Eiffel Tower?"

for cand in candidates:
    input_text = f"question: {question} context: {cand}"
    inputs = tokenizer(input_text, return_tensors="pt")
    labels = tokenizer("Paris", return_tensors="pt").input_ids
    loss = model(**inputs, labels=labels).loss.item()
    print(cand, "→ loss:", loss)
```

```python
import torch
from transformers import T5ForConditionalGeneration, T5Tokenizer

model = T5ForConditionalGeneration.from_pretrained("t5-small")
tokenizer = T5Tokenizer.from_pretrained("t5-small")

question = " is  Eiffel Tower in Paris?"
candidates = [
    "The Eiffel Tower is in Paris.",
    "It was built in 1889."
]

model.eval()
for cand in candidates:
    # Format input
    input_text = f"question: {question} context: {cand}"
    input_ids = tokenizer(input_text, return_tensors="pt").input_ids

    # Tokenize candidate as output
    target_ids = tokenizer(cand, return_tensors="pt").input_ids

    # Shift labels as T5 expects
    labels = target_ids.clone()

    # Forward pass with labels → get loss
    with torch.no_grad():
        outputs = model(input_ids=input_ids, labels=labels)
        loss = outputs.loss.item()

    print(f"Candidate: {cand} → Loss: {loss}")
```

# T5 for Question Answering

```python
from transformers import T5ForConditionalGeneration, T5Tokenizer

model = T5ForConditionalGeneration.from_pretrained("t5-small")
tokenizer = T5Tokenizer.from_pretrained("t5-small")

# Input prompt with task prefix
text = "question answering: What did Joey and Jimmy do at the pond? context: Once upon a time, there was a squirrel named Joey. Joey loved to go outside and play with his cousin Jimmy. One day, Joey and Jimmy went swimming together at their Aunt Julie's pond. They dove into the water and swam for several hours"
inputs = tokenizer(text, return_tensors="pt")

# Generate output
outputs = model.generate(**inputs, max_length=50)
print(tokenizer.decode(outputs[0], skip_special_tokens=True))
```
